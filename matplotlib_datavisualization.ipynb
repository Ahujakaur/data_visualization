# Scater_plot

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5]
no=[6,5,9,8,7]
colors=["r","y","g","b","r"]
sizes=[400,500,200,600,300]
plt.scatter(day,no,c=colors,s=sizes,alpha=0.4,marker="*")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("no",fontsize=15)
plt.show()
```

# pie_plot

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,shadow=True,radius=0.5)
plt.show()
```

# Box_plot

```python
x=[10,20,30,40,50,60]
plt.boxplot(x,notch=True)
plt.show()
```


# Stack plot

```python
x=[1,2,3,4,5]
area1=[2,3,4,5,6]
area2=[2,3,3,2,1]
area3=[2,2,5,8,9]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l)
plt.legend(loc=2)
plt.show()
```

# Step_plot

```python
x=[1,2,3,4,5]
y=[5,6,7,4,5]
plt.title("python")
plt.xlabel("x- axis")
plt.ylabel("y-axis")
plt.grid()
plt.step(x,y)
plt.show()
```
